```python
# Total number of missing values
df.isnull().sum().sum()
```

```python
# Dropping the rows with NaN values, you can use the `how=all` to preserve the row where there are one or two Nan
#  but if all values in the row are NaN then it will drop that row.
df_new = df.dropna()

# Boolean representation to check if the values are values are NaN or not.
df_new = df[df['beds'].notna()]
df_new = df[df['beds'].notna() & df['bedrooms'].notna()]

# noted that df is edited in-place here 
# the df gets edited in-place only when we use inplace=true!
# fillna => fill NaN values.
df['beds'].fillna((df['beds'].mean()), inplace=True)
print(df['beds'])
```

```python
# printing first-10 rows' room types 
print(list(df.room_type)[:10])
```

```python
# encoding
# astype => casts a pandas object to a specified dtype.
# cat => provides with categorical properties of the Series values
# codes => are an array of integers which are the positions of the actual values in the categories array.
encoded_room_type_columne = df.room_type.astype('category').cat.codes
print(list(encoded_room_type_columne)[:10])
```

```python
# replace original:
df.room_type = encoded_room_type_columne
print(list(df.room_type)[:10])
```
